```python
import requests
import json
import pprint

root_url = "http://opendata-download-metobs.smhi.se/api.json"
api = requests.get(root_url).json()
ind1 = next(i for (i, d) in enumerate(api["version"]) if d["key"] == "latest")
ind2 = next(i for (i, d) in enumerate(api["version"][ind1]["link"]) if d["type"] == "application/json")
resources = requests.get(api["version"][ind1]["link"][ind2]["href"]).json()
pprint.pprint(resources)
```
